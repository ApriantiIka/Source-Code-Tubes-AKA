```python
import time
import matplotlib.pyplot as plt
from prettytable import PrettyTable
```

```python
# Iteratif Linear Search
def linear_search(arr, target):
  for i in range(len(arr)):
    if arr[i] == target:
      return i
  return -1
```

```python
# Rekursif Binary Search
def binary_search(arr, target, left, right):
  if left > right:
    return -1
  mid = (left + right) // 2
  if arr[mid] == target:
    return mid
  elif arr[mid] > target:
    return binary_search(arr, target, left, mid - 1)
  else:
    return binary_search(arr, target, mid + 1, right)
```

```python
# Fungsi untuk membuat dataset teks
def generate_dataset(n):
  words = [
    "apple", "banana", "cherry", "date", "fig", "grape", "kiwi", "lemon", "mango", "orange",
    "cat", "dog", "elephant", "frog", "giraffe", "horse", "iguana", "jaguar", "kangaroo", "lion",
    "argentina", "brazil", "canada", "denmark", "egypt", "france", "germany", "hungary", "india", "japan"
  ]

  # Jika n lebih besar dari jumlah kata yang tersedia, ulangi daftar kata
  if n > len(words):
    words = (words * ((n // len(words)) + 1))[:n]
  else:
    words = words[:n]

  # Mengurutkan kata berdasarkan alfabet
  return sorted(words)
```

```python
# Grafik untuk menyimpan data
n_values = []
linear_times = []
binary_times = []

# Fungsi untuk memperbarui grafik
def update_graph():
  plt.figure(figsize=(8, 6))
  plt.plot(n_values, linear_times, label='Linear Search (Iterative)', marker='o', linestyle='-')
  plt.plot(n_values, binary_times, label='Binary Search (Recursive)', marker='o', linestyle='-')
  plt.title('Performance Comparison: Linear Search vs Binary Search')
  plt.xlabel('Input Size (n)')
  plt.ylabel('Execution Time (seconds)')
  plt.legend()
  plt.grid(True)
  plt.show()
```

```python
# Fungsi untuk mencetak tabel waktu eksekusi
def print_execution_table():
  table = PrettyTable()
  table.field_names = ["n", "Linear Search Time (s)", "Binary Search Time (s)"]
  min_len = min(len(n_values), len(linear_times), len(binary_times))
  for i in range(min_len):
    table.add_row([n_values[i], linear_times[i], binary_times[i]])
  print(table)
```

```python
# Program utama
while True:
  try:
    n = int(input("Masukkan ukuran input n (atau ketik -1 untuk keluar): "))
    if n == -1:
      print("Program selesai. Terima kasih!")
      break
    if n <= 0:
      print("Masukkan jumlah elemen yang positif!")
      continue

    # Membuat dataset
    data = generate_dataset(n)
    target = data[n // 2]  # Elemen target ada di tengah untuk pengujian

    n_values.append(n)

    # Ukur waktu eksekusi Linear Search
    start_time = time.time()
    linear_search(data, target)
    linear_times.append(time.time() - start_time)

    # Ukur waktu eksekusi Binary Search
    start_time = time.time()
    binary_search(data, target, 0, n - 1)
    binary_times.append(time.time() - start_time)

    # Cetak tabel waktu eksekusi
    print_execution_table()

    # Perbarui grafik
    update_graph()

  except ValueError:
    print("Masukkan nilai n yang valid!")
```
